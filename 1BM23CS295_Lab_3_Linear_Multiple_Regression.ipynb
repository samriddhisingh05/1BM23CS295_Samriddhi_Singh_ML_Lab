```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

print(df.head())

# Split data
X = df[['year']]
y = df['per capita income (US$)']

# Train model
model = LinearRegression()
model.fit(X, y)

# Prediction for 2020
prediction = model.predict([[2020]])

print("Predicted income for 2020:", prediction[0])
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

x = df['year'].values
y = df['per capita income (US$)'].values

n = len(x)

# Calculate slope (m)
m = (n*sum(x*y) - sum(x)*sum(y)) / (n*sum(x*x) - (sum(x))**2)

# Calculate intercept (b)
b = (sum(y) - m*sum(x)) / n

print("Slope:", m)
print("Intercept:", b)

# Predict for 2020
year = 2020
predicted = m*year + b

print("Predicted income for 2020:", predicted)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Load dataset
df = pd.read_csv("salary.csv")

# Define X and y
X = df[['YearsExperience']]
y = df['Salary']

# Create imputer (replace NaN with mean)
imputer = SimpleImputer(strategy='mean')

X = imputer.fit_transform(X)

# Build model
model = LinearRegression()
model.fit(X, y)

# Predict for 12 years
prediction = model.predict([[12]])

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
print("Predicted Salary for 12 years:", prediction[0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("salary.csv")

X = df['YearsExperience'].values
y = df['Salary'].values

# Replace NaN with mean manually
mean_value = np.nanmean(X)
X = np.where(np.isnan(X), mean_value, X)

# Add column of 1s
X = np.column_stack((np.ones(len(X)), X))

# Normal Equation
beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print("Intercept:", beta[0])
print("Slope:", beta[1])

# Predict for 12 years
X_new = np.array([1, 12])
prediction = X_new.dot(beta)

print("Predicted Salary for 12 years:", prediction)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Load dataset
df = pd.read_csv("hiring.csv")

print(df.head())

# Convert text numbers to numeric
mapping = {
    'zero':0, 'one':1, 'two':2, 'three':3,
    'four':4, 'five':5, 'six':6,
    'seven':7, 'eight':8, 'nine':9, 'ten':10
}

df['experience'] = df['experience'].map(mapping)

# Convert to numeric (in case of mixed types)
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

# Define X and y
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Build model
model = LinearRegression()
model.fit(X, y)

print("\nIntercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Predictions
print("\nSalary for (2,9,6):", model.predict([[2,9,6]])[0])
print("Salary for (12,10,10):", model.predict([[12,10,10]])[0])
```

```python
import numpy as np

df = pd.read_csv("hiring.csv")

# Convert text to numeric
mapping = {
    'zero':0, 'one':1, 'two':2, 'three':3,
    'four':4, 'five':5, 'six':6,
    'seven':7, 'eight':8, 'nine':9, 'ten':10
}

df['experience'] = df['experience'].map(mapping)
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']].values
y = df['salary($)'].values

# Replace NaN with column mean
col_means = np.nanmean(X, axis=0)
inds = np.where(np.isnan(X))
X[inds] = np.take(col_means, inds[1])

# Add intercept column
X = np.column_stack((np.ones(len(X)), X))

# Normal Equation
beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print("\nIntercept:", beta[0])
print("Coefficients:", beta[1:])

# Predictions
new1 = np.array([1,2,9,6])
new2 = np.array([1,12,10,10])

print("\nSalary for (2,9,6):", new1.dot(beta))
print("Salary for (12,10,10):", new2.dot(beta))
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("1000_Companies.csv")

# Convert State column into dummy variables
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Convert all columns to float (VERY IMPORTANT FIX)
df = df.astype(float)

# Input variables
X = df.drop('Profit', axis=1).values

# Output variable
Y = df['Profit'].values

# Add intercept column (1s)
ones = np.ones((X.shape[0], 1))
X = np.concatenate((ones, X), axis=1)

# Apply Normal Equation
B = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)

print("Coefficients:", B)

# Prediction for given input
# Florida = State_Florida=1, State_New York=0
x_test = np.array([1, 91694.48, 515841.3, 11931.24, 1, 0], dtype=float)

prediction = x_test.dot(B)

print("\nWithout Library:")
print("Predicted Profit:", prediction)
```
